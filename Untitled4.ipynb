```python
# single inheritance between two classes ; library and book
class Library:
    def __init__(self, book_id):
        self.book_id = book_id


class Book(Library):
    def __init__(self, book_id, author_name, book_name):
        super().__init__(book_id)
        self.author_name = author_name
        self.book_name = book_name

    def display(self):
        print("Book ID :", self.book_id)
        print("Author Name :", self.author_name)
        print("Book Name :", self.book_name)


book1 = Book(101, "Bhagat Singh", "Why I am an Atheist")

book1.display()
```

```python
# vehicle rental system
class Vehicle:
    def __init__(self,name):
        self.name = name

class Bus(Vehicle):
    def calculate_rent(self, km):
        rent = 7500
        if km > 250:
            rent += (km - 250) * 70
        return rent

class Car(Vehicle):
    def calculate_rent(self, km):
        rent = 5000
        if km > 250:
            rent += (km - 250) * 50
        return rent

class Bike(Vehicle):
    def calculate_rent(self, km):
        rent = 2000
        if km > 250:
            rent += (km - 250) * 20
        return rent



bus = Bus("Volvo Bus")
car = Car("Swift")
bike = Bike("Activa")

km = int(input("Enter distance travelled (km): "))

print("Bus Rent =", bus.calculate_rent(km))
print("Car Rent =", car.calculate_rent(km))
print("Bike Rent =", bike.calculate_rent(km))
```

```python
# generate a salary slip if you are HR having employees : manager, developer, intern
class Employee:
    def __init__(self, name, designation, basic_salary, overtime_rate):
        self.name = name
        self.designation = designation
        self.basic_salary = basic_salary
        self.overtime_rate = overtime_rate

    def generate_salary_slip(self, extra_hours):
        overtime_pay = extra_hours * self.overtime_rate
        total_salary = self.basic_salary + overtime_pay

        print("\n------ SALARY SLIP ------")
        print("Employee Name :", self.name)
        print("Designation   :", self.designation)
        print("Basic Salary  :", self.basic_salary)
        print("Extra Hours   :", extra_hours)
        print("Overtime Pay  :", overtime_pay)
        print("Total Salary  :", total_salary)
        print("-------------------------\n")


print("1. Manager")
print("2. Developer")
print("3. Intern")

choice = int(input("Enter employee category: "))
name = input("Enter employee name: ")
extra_hours = int(input("Enter extra hours worked: "))

if choice == 1:
    emp = Employee(name, "Manager", 80000, 800)

elif choice == 2:
    emp = Employee(name, "Developer", 50000, 500)

elif choice == 3:
    emp = Employee(name, "Intern", 15000, 150)

else:
    print("Invalid Choice")
    exit()

emp.generate_salary_slip(extra_hours)
```

```python
class ElectricityBill:
    def __init__(self):
        self.devices = []

    def get_units(self):
        print("Enter electricity consumption of 8 devices:")
        for i in range(1, 9):
            units = float(input(f"Device {i} units: "))
            self.devices.append(units)

    def calculate_bill(self):
        total_units = sum(self.devices)

        if total_units <= 300:
            bill = total_units * 10
        elif total_units <= 600:
            bill = total_units * 15
        else:
            bill = total_units * 20

        print("\n----- ELECTRICITY BILL -----")
        print("Total Units Consumed:", total_units)
        print("Total Bill Amount: Rs.", bill)


obj = ElectricityBill()
obj.get_units()
obj.calculate_bill()7
```
